### Seaborn Color Palettes

Seaborn provides several built-in color palettes that make it easy to create visually appealing plots. Here, we'll cover three types of color palettes: Qualitative, Sequential, and Diverging.

#### Qualitative Color Palettes
Qualitative color palettes are used to represent categorical data where there is no inherent order. Popular qualitative color palettes include `Set2` and `Paired`.

```python
# Example data
categories = ['Holland', 'France', 'Spain', 'Portugal', 'England']
values = [8, 10, 20, 15, 1]

# Create a DataFrame
data = pd.DataFrame({'Category': categories, 'Value': values})

# Set color palette
palette = sns.color_palette("Set2", len(categories))

# Create a bar plot with different hues 
sns.barplot(x='Category', y='Value', data=data, hue='Category', palette=palette)
plt.title("Goals Scored By Each Team")
plt.xlabel("Teams")
plt.ylabel("Total Goals Scored")
plt.show()
```

#### Sequential Color Palettes
Sequential color palettes are used to represent data that has a natural ordering, such as values from low to high. Popular sequential color palettes include `flare`, `mako` and `Blues`.

```python
# Generate example data for 100 cities
np.random.seed(0)
cities = [f'City {i}' for i in range(1, 101)]
temperatures = np.random.uniform(10, 30, 100)  
populations = np.random.uniform(100000, 10000000, 100)  

# Create a DataFrame
data = {
    'City': cities,
    'Temperature': temperatures,
    'Population': populations
}
df = pd.DataFrame(data)

# Set the color palette
palette = sns.color_palette("Blues", as_cmap=True)

# Create a scatter plot with a sequential color palette
sns.scatterplot(x='Temperature', y='Population', hue='Temperature', palette=palette, data=df, legend = False)
plt.title("City Temperatures vs Population")
plt.xlabel("Average Annual Temperature (Â°C)")
plt.ylabel("Population")
plt.show()
```

#### Diverging Color Palettes
Diverging color palettes are used to represent data that has a meaningful midpoint, such as zero, with values diverging in both directions. Popular diverging color palettes include `icefire` and `vlag`.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Example data: Average monthly temperatures (Â°C) for 10 cities
data = np.array([
    [-2, 3, 5, 9, 13, 18, 20, 19, 15, 10, -5, -3],
    [1, 2, 5, 8, 12, 17, 19, 18, 14, 9, -4, -2],
    [0, 1, 4, 7, 11, 16, 18, 17, 13, 8, -3, -1],
    [-3, 5, 8, 12, 16, 21, 23, 22, 18, 13, -7, -4],
    [-5, 7, 10, 14, 18, 23, 25, 24, 10, 4, -10, -6],
])

# City and month labels
cities = ["Lahore", "New York", "Paris", "Tokyo", "Sydney"]
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

# Convert the data to a DataFrame
df = pd.DataFrame(data, index=cities, columns=months)

# Set the color palette
sns.set_palette("coolwarm")

# Create a heatmap
sns.heatmap(df, center=0, cmap="icefire")
plt.title("Heatmap of Average Monthly Temperatures for 10 Cities")
plt.xlabel("Month")
plt.ylabel("City")
plt.show()
```
